```python
# SK88 Create a treemap visualization for different categories of eBay for average bid amounts and negative feedback rates
# SK88 import the required packages : squarify and matplotlib

import squarify
import matplotlib

ebayTreemapSK88 = dmba.load_data('EbayTreemap.csv')

groupedSK88 = []
for categorySK88, dfSK88 in ebayTreemapSK88.groupby('Category'):
    negativeFeedbackSK88 = sum(dfSK88['Seller Feedback'] < 0) / len(dfSK88)
    groupedSK88.append({
        'category': categorySK88,
        'negativeFeedback': negativeFeedbackSK88,
        'averageBid': dfSK88['High Bid'].mean()
    })
byCategorySK88 = pd.DataFrame(groupedSK88)

normSK88 = matplotlib.colors.Normalize(vmin=byCategorySK88.negativeFeedback.min(), vmax=byCategorySK88.negativeFeedback.max())
colorsSK88 = [matplotlib.cm.Blues(normSK88(value)) for value in byCategorySK88.negativeFeedback]

figSK88, axSK88 = plt.subplots()
figSK88.set_size_inches(9, 5)

renameCategoriesSK88 = {
    'Business & Industrial': 'Business &\nIndustrial',
    'Health & Beauty': 'Health &\nBeauty',
    'Consumer Electronics': 'Consumer\nElectronics',
    'Clothing & accessories': 'Clothing &\naccessories',
    'Clothing shoes & accessories': 'Clothing shoes &\naccessories'
}
labelsSK88 = [renameCategoriesSK88.get(cSK88, cSK88) for cSK88 in byCategorySK88.category]

squarify.plot(label=labelsSK88, sizes=byCategorySK88.averageBid, color=colorsSK88,
              ax=axSK88, alpha=0.6, edgecolor='grey', text_kwargs={'fontsize': 8.7})

axSK88.get_xaxis().set_ticks([])
axSK88.get_yaxis().set_ticks([])

plt.subplots_adjust(left=0.1)
plt.show()
# labels
```

```python
# SK88 Create a variable 'gdp_dfSK88' to load GDP and happiness data for the countries
# SK88 Create a visualiztion of world map using the package Cartopy

import matplotlib
import matplotlib.pyplot as plt
import cartopy
import cartopy.io.shapereader as shpreader
import cartopy.crs as ccrs

gdp_dfSK88 = dmba.load_data('gdp.csv', skiprows=4)
gdp_dfSK88 = gdp_dfSK88.rename(columns={'2015': 'GDP2015'})
gdp_dfSK88 = gdp_dfSK88.set_index('Country Code') # use the three letter country code to access rows

# SK88 The file contains a column with two letter combinations, use na_filter to avoid converting
# SK88 The combination NA into not-a-number
happiness_dfSK88 = dmba.load_data('Veerhoven.csv', na_filter=False)
happiness_dfSK88 = happiness_dfSK88.set_index('Code')  # use the country name to access rows


figSK88, (ax1SK88, ax2SK88) = plt.subplots(2, 1, figsize=(7, 8), subplot_kw={"projection": ccrs.PlateCarree()})
ax1SK88.set_extent([-150, 60, -25, 60])
ax2SK88.set_extent([-150, 60, -25, 60])

# SK88 Create a color mapper
cmapSK88 = plt.cm.Blues_r
norm1SK88 = matplotlib.colors.Normalize(vmin=happiness_dfSK88.Score.dropna().min(),
                                        vmax=happiness_dfSK88.Score.dropna().max())
norm2SK88 = matplotlib.colors.LogNorm(vmin=gdp_dfSK88.GDP2015.dropna().min(),
                                      vmax=gdp_dfSK88.GDP2015.dropna().max())

shpfilenameSK88 = shpreader.natural_earth(resolution='110m',
                                           category='cultural',
                                           name='admin_0_countries')
readerSK88 = shpreader.Reader(shpfilenameSK88)
countriesSK88 = readerSK88.records()
for countrySK88 in countriesSK88:
    countryCodeSK88 = countrySK88.attributes['ADM0_A3']
    if countryCodeSK88 in gdp_dfSK88.index:
        if pd.isna(gdp_dfSK88.loc[countryCodeSK88].GDP2015):
            ax2SK88.add_geometries([countrySK88.geometry], ccrs.PlateCarree(),
                                   edgecolor='grey', facecolor='white')
        else:
            ax2SK88.add_geometries([countrySK88.geometry], ccrs.PlateCarree(),
                                   facecolor=cmapSK88(norm2SK88(gdp_dfSK88.loc[countryCodeSK88].GDP2015)))
    # SK88 Check various attributes to find the matching two-letter combinations
    nationSK88 = countrySK88.attributes['POSTAL']
    if nationSK88 not in happiness_dfSK88.index:
        nationSK88 = countrySK88.attributes['ISO_A2']
    if nationSK88 not in happiness_dfSK88.index:
        nationSK88 = countrySK88.attributes['WB_A2']
    if nationSK88 not in happiness_dfSK88.index and countrySK88.attributes['NAME'] == 'Norway':
        nationSK88 = 'NO'
    if nationSK88 in happiness_dfSK88.index:
        ax1SK88.add_geometries([countrySK88.geometry], ccrs.PlateCarree(),
                               facecolor=cmapSK88(norm1SK88(happiness_dfSK88.loc[nationSK88].Score)))

ax2SK88.set_title("GDP 2015")
smSK88 = plt.cm.ScalarMappable(norm=norm2SK88, cmap=cmapSK88)
cbSK88 = plt.colorbar(smSK88, ax=ax2SK88)
cbSK88.set_ticks([1e8, 1e9, 1e10, 1e11, 1e12, 1e13])

ax1SK88.set_title("Happiness")
smSK88 = plt.cm.ScalarMappable(norm=norm1SK88, cmap=cmapSK88)
cbSK88 = plt.colorbar(smSK88, ax=ax1SK88)
cbSK88.set_ticks([3, 4, 5, 6, 7, 8])

plt.show()
```
